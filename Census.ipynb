```python
import os, glob
import pandas as pd
import csv
#pull in all census csv fiiles and merge into one df
path = "census education data"

all_files = glob.glob(os.path.join(path, "census_data*.csv"))
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
df_merged   = pd.concat(df_from_each_file, ignore_index=True)
#look at the original column header(s) and use the headers originally in row 2 of the csv file
df_merged.columns=list(df_merged.iloc[0])
#once headers have been updated, remove(drop) row 2, which has index of 0
df_merged=df_merged.drop([0])
#drop rows with NaN
df_merged=df_merged.dropna()
# split the Geographic Area Name column into two columnes for county and state 
df_merged[['County','State']] = df_merged['Geographic Area Name'].str.split(",",expand=True) 
#create new csv file with the merged and newly formatted data
df_merged.to_csv( "merged.csv")
df_merged
```
